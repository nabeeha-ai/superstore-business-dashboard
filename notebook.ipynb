# Task 5: Interactive Business Dashboard in Streamlit — Data Preparation & EDA

## 1. Problem Statement and Objective
Business stakeholders need a quick, interactive way to analyze **sales and profit performance** across regions, categories, and customers, without waiting on ad-hoc analyst reports.

**Objective:** Clean and explore a Global-Superstore-style sales dataset, then build an interactive **Streamlit dashboard** (`app.py`) with filters for Region / Category / Sub-Category and KPIs for Total Sales, Profit, and Top 5 Customers by Sales.

## 2. Dataset Description
The dataset mirrors the **Global Superstore** dataset structure: `Order ID`, `Order Date`, `Region`, `Category`, `Sub-Category`, `Customer Name`, `Quantity`, `Unit Price`, `Discount`, `Sales`, `Profit`.

> Note: Since this environment cannot reach external dataset hosts, a synthetic dataset (`data/global_superstore.csv`) of 3,000 orders was generated with realistic category-level pricing, discounting, and margin patterns (e.g. Technology has higher margins than Furniture, higher discounts erode profit). This notebook covers the analysis; `app.py` is the actual interactive dashboard — run it with `streamlit run app.py`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

## 3. Load and Explore the Dataset

```python
df = pd.read_csv("data/global_superstore.csv", parse_dates=["Order Date"])
print(df.shape)
df.head()
```

```python
df.info()
print()
print(df.isna().sum())
```

## 4. Data Cleaning & Preparation

```python
# Drop exact duplicate orders, ensure numeric types, add derived time features
df = df.drop_duplicates(subset="Order ID")
df["Sales"] = df["Sales"].astype(float)
df["Profit"] = df["Profit"].astype(float)
df["Profit Margin"] = (df["Profit"] / df["Sales"]).round(3)
df["Order Month"] = df["Order Date"].dt.to_period("M").astype(str)
df["Order Year"] = df["Order Date"].dt.year
df.describe()
```

## 5. Exploratory Data Analysis

```python
fig, ax = plt.subplots()
df.groupby("Region")["Sales"].sum().sort_values().plot(kind="barh", ax=ax, color="#4C72B0")
ax.set_title("Total Sales by Region")
ax.set_xlabel("Sales ($)")
plt.show()
```

```python
fig, ax = plt.subplots()
df.groupby("Category")[["Sales","Profit"]].sum().plot(kind="bar", ax=ax)
ax.set_title("Sales vs Profit by Category")
plt.xticks(rotation=0)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
monthly = df.groupby("Order Month")["Sales"].sum().sort_index()
monthly.plot(ax=ax, marker="o", color="#55A868")
ax.set_title("Monthly Sales Trend")
ax.set_ylabel("Sales ($)")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(5)
fig, ax = plt.subplots()
top_customers.sort_values().plot(kind="barh", ax=ax, color="#C44E52")
ax.set_title("Top 5 Customers by Sales")
ax.set_xlabel("Sales ($)")
plt.show()
top_customers
```

```python
fig, ax = plt.subplots()
sns.boxplot(data=df, x="Category", y="Profit Margin", ax=ax)
ax.set_title("Profit Margin Distribution by Category")
ax.axhline(0, color="red", linestyle="--", linewidth=1)
plt.show()
```

```python
# Save the cleaned dataset used by the dashboard app
df.to_csv("data/global_superstore_clean.csv", index=False)
print("Saved cleaned dataset:", df.shape)
```

## 6. Conclusion & Insights
- **Furniture** generates solid sales volume but the **thinnest profit margins** (heavy discounting on tables/bookcases erodes profit, occasionally into losses).
- **Technology** delivers the **highest profit margins**, making it the most valuable category per dollar of sales — worth prioritizing in upsell/cross-sell campaigns.
- The **West and East regions** are the strongest sales contributors; underperforming regions may need localized promotions or better assortment.
- A small group of **top 5 customers** contributes disproportionately to total sales — an account-management or loyalty program targeting them could protect and grow this revenue base.
- Monthly sales show clear fluctuation, useful for future demand-planning and staffing decisions.

The companion `app.py` turns these findings into a live, filterable **Streamlit dashboard** so stakeholders can self-serve this analysis by Region / Category / Sub-Category.
